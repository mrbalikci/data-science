```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('data/OnlineRetail.csv', encoding = 'unicode_escape')
df.head()
```

```python
df.Description = df.Description.str.strip()
```

```python
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True) # drop rows that do not have invoice numbers
```

```python
df['InvoiceNo'] = df.InvoiceNo.astype('str')
df = df[~df.InvoiceNo.str.contains('C')]
df.head()
```

```python
basket = (df[df.Country=='France']
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
basket.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >=1:
        return 1
```

```python
basket_sets = basket.applymap(encode_units)
basket_sets.drop('POSTAGE', inplace=True, axis=1)
basket_sets.head()
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rules.head()
```

```python
# with high confidence, how the products should be rearranged 
# with association rule 
rules[(rules.lift >=6) & (rules.confidence >=0.8)]
```
